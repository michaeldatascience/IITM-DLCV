## Pixel spacing $\Delta$

### What does $\Delta$ represent physically?

Physically, $\Delta$ (delta) represents the **physical dimension corresponding to one pixel in the real world**. It is the size of the area in the scene that a single sensor element (photosite) captures. For example, if $\Delta$ is 1 micrometer, it means each pixel in the digital image represents a 1 micrometer by 1 micrometer square in the scene being imaged.

## Three operation types

| Operation | Pixels used | Cost per pixel | Total cost (N×N image) | Example |
|:---|:---|:---|:---|:---|
| **Point** | Only (i,j) itself | O(1) | O(N²) | Contrast inversion, brightness +50 |
| **Local** | p×p neighborhood around (i,j) | O(p²) | O(N²p²) | Gaussian blur, Sobel edge detection |
| **Global** | All N×N pixels | O(N²) | O(N⁴) naive | Histogram equalization, Fourier transform |

## Three point operations

Point operations modify the intensity of each pixel independently, based only on its original value.

### 1. Contrast Inversion

$Ĩ[i,j] = I_{MAX} - I[i,j] + I_{MIN} $

Where:
*   $Ĩ$ is the output pixel value.
*   $I_{max}$ is the maximum possible intensity value (e.g., 255 for an 8-bit image). and $I_{MIN}$ is minimum.
*   $I$ is the input pixel value.

### 2. Linear Contrast Stretching

$Ĩ[i,j] = \frac{I[i,j]-I_{MIN}}{I_{MAX} - I_{MIN}} \times (I_{MAX} - I_{MIN}) + I_{MIN} $



Rounding to the nearest integer, the output pixel value would be $\approx 128$.

### 3. Histogram Equalization

**Concept:** This process is to qualize and flatten the image's intensity, distributing the intensity values uniformally across the complete range.
This generally results in enhanced contrast of features that were previously hidden in regions of low contrast.

**Formula (for discrete images):**

$$
c_{k} =  \frac{1}{M \times N} ∑_{i=0}^{k} h(l)
$$

$c_{k}$ ⟶ fraction of all pixels having density ≤ k

$$
Ĩ[i,j] = I_{MAX} \times c_{l}[i,j]
$$


## Outlier problem in linear stretch

A single outlier pixel with an extremely high or low intensity value can severely compress the usable dynamic range for the majority of the image during linear contrast stretching, making the image still appear dark or washed out. The fix for this is **percentile clipping**, where the top and bottom few percent of pixel values are clipped (i.e., mapped to $O_{max}$ or $O_{min}$) before applying the linear stretch, effectively ignoring the outliers and allowing the stretch to focus on the significant portion of the image's intensity range.
